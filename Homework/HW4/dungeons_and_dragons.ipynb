```python
import random
import matplotlib.pyplot as plt
```

1a) Write a Python function that rolls a die with num_sides sides a total of
num_rolls times. The function should plot a histogram of all of the rolls and
should return the average roll value.

```python
def roll_dice(num_sides, num_rolls):
    dict = {}
    # roll the dice
    i = 0
    while i < num_rolls:
        random_number = random.randint(1, num_sides)
        if random_number in dict:
            dict[random_number] += 1
        else:
            dict[random_number] = 1
        i+=1
        
    x_values = list(dict.keys())
    y_values = list(dict.values())
    
    print(x_values)
    print(y_values)
    
    #plot histogram
    plt.bar(x_values, y_values, width=1, align='center', color='blue', edgecolor='black')
    plt.xticks(x_values)
    plt.xlabel('Sides on Die')
    plt.ylabel('Frequency of Roll')
    plt.title('Histogram of All Roles of num_sides-Sided Die')
    plt.show()
    
    #caluclate average value rolled
    total = 0
    for key, value in dict.items():
        total += (key*value)

    return total/num_rolls
```

1b) Use this function to simulate rolling a 20-sided die 1000000 times, plot a
histogram of the results, and calculate the average roll value. Is the average
roll value what you would expect? Comment on the shape of the histogram.

```python
avg_roll = roll_dice(20,1000000)
print("The average value rolled on a 20-sided die that was rolled 1000000 times is:", avg_roll)
```

The average roll value expected when rolling a 20-sided die 1000000 times is calculated as follows:

=(1+2+3+...+19+20)/20
=210/20
=10.5

The average value we calculated from our experiment was about 10.5.  Thus, the value we calculated from our experiment was expected!  Lastly, the shape of the histogram is fairly even.  This means that as we rolled the die more and more times the number of times specific sides were rolled began to even out, which makes sense.  

1c) Write a Python function that rolls and n-sided die a total of m times, and
returns the MAXIMUM value rolled.

```python
def get_max(n_sides, m_times):
    i = 0
    max = 0
    while i < m_times:
        random_number = random.randint(1, n_sides)
        if random_number > max:
            max = random_number
        i += 1
    return max
```

1d) Simulate rolling a 20-sided die, where one rolls the die TWICE, and takes the
maximum of the two rolls. Plot a histogram of the results of doing this
1000000. Does the shape of the histogram make sense?

```python
dict = {}

# populate the dict
i = 0
while i < 1000000:
    max = get_max(20,2)
    if max in dict:
        dict[max] += 1
    else:
        dict[max] = 1
    i += 1
    
# plot the histogram

x_values = list(dict.keys())
y_values = list(dict.values())
    
print(x_values)
print(y_values)

plt.bar(x_values, y_values, width=1, align='center', color='blue', edgecolor='black')
plt.xticks(x_values)
plt.xlabel('Sides on Die')
plt.ylabel('Frequency of Roll')
plt.title('Histogram of Max Role of 20-Sided Die Rolled Twice')
plt.show()
```

The shape of the histogram does make sense.  20 is the maximum value you can roll on a 20-sided die, so it makes sense that after a plethora of rolls, this value would come out to be the max value rolled.  I was a little surprised about the steady increase at first, but thinking more about it, it does make sense.  1 is very rarely going to be the max value rolled when a 20-sided die is rolled twice unless on both rolls a 1 is rolled.  Similarly, 2 is very rarely going to be the max value rolled using similar logic as before.  Continuing with this logic, if a 20 is ever rolled, it will be the max value, which makes sense why is has the most counts for being the max value rolled.  Thus, the shape of the histogram makes sense.
